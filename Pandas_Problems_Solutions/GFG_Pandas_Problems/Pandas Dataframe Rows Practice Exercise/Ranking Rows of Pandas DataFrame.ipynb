```python
import pandas as pd  

movies = {'Name': ['The Godfather', 'Bird Box', 'Fight Club'], 
         'Year': ['1972', '2018', '1999'], 
         'Rating': ['9.2', '6.8', '8.8']} 
  
df = pd.DataFrame(movies) 
df
```

### df.rank()-->computes numerical rank along the axis, if tie between ranks then mean of ranks is assigned,rank in ascending order

```python
df['Rating_Rank']=df['Rating'].rank()
df
```

```python
df.set_index('Rating_Rank').sort_index()
```

```python
 student_details = {'Name':['Raj', 'Raj', 'Raj', 'Aravind', 'Aravind', 'Aravind', 
                             'John', 'John', 'John', 'Arjun', 'Arjun', 'Arjun'], 
                   'Subject':['Maths', 'Physics', 'Chemistry', 'Maths', 'Physics', 
                             'Chemistry', 'Maths', 'Physics', 'Chemistry', 'Maths', 
                                                           'Physics', 'Chemistry'], 
                   'Marks':[80, 90, 75, 60, 40, 60, 80, 55, 100, 90, 75, 70] 
               } 
df = pd.DataFrame(student_details) 
df
```

### We will rank the students based on the highest mark they have scored.

```python
df['Mark_Rank']=df['Marks'].rank(ascending=False)
df.set_index('Mark_Rank').sort_index()
```

```python
df['Mark_Rank']=df['Marks'].rank(method='max',ascending=False)
df.set_index('Mark_Rank').sort_index()
```
