```python
import pandas as pd
```

```python
record = {
 'Name': ['Ankit', 'Amit', 'Aishwarya', 'Priyanka', 'Priya', 'Shaurya' ],
 'Age': [21, 19, 20, 18, 17, 21],
 'Stream': ['Math', 'Commerce', 'Science', 'Math', 'Math', 'Science'],
 'Percentage': [88, 92, 95, 70, 65, 78] }

# create a dataframe
dataframe = pd.DataFrame(record, columns = ['Name', 'Age', 'Stream', 'Percentage'])

#print(&quot;Given Dataframe :\n&quot;, dataframe)
dataframe
```

#### Selecting all the rows from the given dataframe in which ‘Percentage’ is greater than 80 using basic method.

```python
dataframe[dataframe['Percentage']>80]
```

#### using loc[]

```python
dataframe.loc[dataframe['Percentage']>80]
```

#### Selecting all the rows from the given dataframe in which ‘Percentage’ is not equal to 95 using Loc.

```python
dataframe
```

```python
dataframe.loc[dataframe['Percentage']!=95]
```

#### Selecting all the rows from the given dataframe in which ‘Stream’ is present in the options list using isin() method.

```python
dataframe
```

```python
dataframe[dataframe.Stream.isin(['Math','Science'])]
```

#### Selecting all the rows from the given dataframe in which ‘Stream’ is present in the options list using .loc[].

```python
dataframe.loc[dataframe.Stream.isin(['Math','Science'])]
```

#### Selecting all the rows from the given dataframe in which ‘Stream’ is not present in the options list using .loc[].

```python
dataframe.loc[~dataframe.Stream.isin(['Math','Science'])]
```

#### Selecting all the rows from the given dataframe in which ‘Age’ is equal to 21 and ‘Stream’ is present in the options list using basic method.

```python
dataframe[(dataframe['Age']==21) & (dataframe['Stream'].isin(['Math','Science']))]
```

#### Selecting all the rows from the given dataframe in which ‘Age’ is equal to 21 and ‘Stream’ is present in the options list using
.loc[].

```python
dataframe.loc[(dataframe['Age']==21) & (dataframe['Stream'].isin(['Math','Science']))]
```
