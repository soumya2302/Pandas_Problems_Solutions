#### drop()--> drop specified label from row or columns

```python
df=pd.read_csv('../DataSetGFG/nba.csv')
df.head()
```

#### Dataframe only contains those players whose age is greater than or equal to 25 years.

```python
filtered_df=df[df['Age']>=25]
filtered_df.head()
```

#### the returned Dataframe only contains those players whose age is not between 20 to 25 age using df.drop(). 

```python
dataset=df[(df['Age']>=20) & (df['Age']<=25)].index
df.drop(dataset).head(16)
```

####  we drop all the rows whose names and Positions are associated with ‘John Holland‘ or ‘SG’ using df.drop().

```python
df.head()
```

```python
data=df[(df['Name']=='John Holland') | (df['Position']=='SG')].index
data
```

```python
df.drop(labels=data).head(10)
```

### Using query()-->
query columns of dataframe with bool expression, drop rows where expression evaluates to False or oonly return rows where bool expression evaluates to True

#### work like boolean indexing

#### keep only individuals aged 30 or younger

```python
data = {'name': ['Alice', 'Bob', 'Charlie', 'Dave'],
        'age': [25, 30, 35, 40],
        'gender': ['F', 'M', 'M', 'M']}
df = pd.DataFrame(data)
df
```

```python
df.query('age <= 30') #====>df[df['age']<=30]
```

## df.loc[]

### retain rows where the ‘age’ is 30 or less,

```python
df
```

```python
df.loc[df['age']<=30]
```

### df.isin(values)

```python
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
        'Age': [25, 30, 35, 40, 45],
        'Score': [85, 92, 78, 65, 89]}
 
df = pd.DataFrame(data)
df
```

#### generates a DataFrame with ‘Name,’ ‘Age,’ and ‘Score,’ filters out rows where ‘Score’ is less than 80

```python
df[df['Score']<80]
```
