```python
new_series=pd.isna(employees['Gender'])
employees[new_series] #bollean indexing wherever condition satisfiees gives True return those rows

# only the rows having Gender = NULL are displayed
```

#### In order to check null values in Pandas Dataframe, we use notnull() function this function return dataframe of Boolean values which are False for NaN values.

```python
employees.notnull()
```

#### only the rows having Gender = NOT NULL are displayed.

```python
data=pd.notnull(employees['Gender'])
employees[data]                
```

#### Filling missing values using fillna(), replace() and interpolate()

#### df.fillna()-->fill df's missing values with specified value,Filling null values with a single value 

```python
employees.fillna(0)
```

#### Filling null values with the previous ones 

```python
employees.fillna(method='pad').head(4)
```

#### filling null value with the next ones 

```python
employees.fillna(method='bfill').head(5)
```

```python
employees.head()
```

```python
employees['Team'].fillna('No Team',inplace=True)
employees
```

#### Filling a null values using replace() method 

```python
employees.replace(to_replace = np.nan, value = -99) 
employees.head()
```
