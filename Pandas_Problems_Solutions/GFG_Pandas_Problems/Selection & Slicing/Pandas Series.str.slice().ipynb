```python
import pandas as pd
df=pd.read_csv('../DataSetGFG/nba.csv')
df.head()
```

##### the salary column has been sliced to get values before decimal. For example, we want to do some mathematical operations and for that we need integer data, so the salary column will be sliced till the 2nd last element(-2 position)

```python
df.Salary=df.Salary.astype('str')
```

```python
df.Salary.dtypes
```

```python
df.dtypes
```

```python
df["Salary (int)"]=df['Salary'].str.slice(stop=-2)
```

##### the name column is sliced and step parameter is kept 2

```python
df.Name=df.Name.str.slice(stop=-2,step=2)
```
