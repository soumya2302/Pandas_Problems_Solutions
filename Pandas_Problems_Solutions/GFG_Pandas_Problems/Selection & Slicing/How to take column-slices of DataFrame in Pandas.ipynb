```python
import pandas as pd
 
# Using DataFrame() method from pandas module
df1 = pd.DataFrame({"a": [1, 2, 3, 4, 5, 6, 7], 
                    "b": [2, 3, 4, 2, 3, 4, 5], 
                    "c": [3, 4, 5, 2, 3, 4, 5], 
                    "d": [4, 5, 6, 2, 3, 4, 5], 
                    "e": [5, 6, 7, 2, 3, 4, 5]})
df1
```

### Slice Columns in pandas using reindex

```python
df1.reindex(['d','a'],axis=1)
```

### Slice Columns in pandas using loc[]

```python
df1.loc[:,'b':'d':2]
```

#### Slicing column from ‘c’ to ‘e’ with step 1.

```python
df1.loc[:,'c':'e']
```

### Slice Columns in pandas using iloc[]

```python
df1.iloc[:,0:3:1]
```
