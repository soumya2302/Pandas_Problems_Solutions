```python
import pandas as pd 
  
# Define a dictionary containing ICC rankings 
rankings = {'test': ['India', 'South Africa', 'England', 
                            'New Zealand', 'Australia'], 
            'odi': ['England', 'India', 'New Zealand', 
                            'South Africa', 'Pakistan'], 
            't20': ['Pakistan', 'India', 'Australia', 
                            'England', 'New Zealand']} 
  
# Convert the dictionary into DataFrame 
rankings_pd = pd.DataFrame(rankings)
rankings_pd
```

### Using rename() function

#### Rename a single column.

```python
rankings_pd.rename(mapper={'test':'Test'},axis=1)
```

#### Rename multiple columns. 

```python
rankings_pd.rename(mapper={'test':'Test','odi':'Odi','t20':'T20'},axis=1)
```

```python
rankings_pd.rename(lambda x:x.title(),axis=1)
```

### By assigning a list of new column names 

```python
rankings_pd.columns=['Test','Odi','T20']
```

```python
rankings_pd
```

```python
rankings_pd
```

```python
rankings_pd.columns.values[0]='test'
```

### Rename column names using DataFrame set_axis() function

```python
rankings_pd.set_axis([11,12,13,14,15])
```

```python
rankings_pd.set_axis(['TEST','ODI','t20'],axis=1)
```

### Rename column names using DataFrame add_prefix() and add_suffix() functions

##### we will pass the prefix and suffix that should be added to the first and last name of the column name.

```python
rankings_pd.add_prefix('row_')
```

```python
rankings_pd.add_prefix('row_',axis=0)
```

```python
rankings_pd.add_suffix('row_')
```

```python
rankings_pd.add_suffix('row_',axis=0)
```
